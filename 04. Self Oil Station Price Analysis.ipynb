## 4. 데이터 정리하기

```python
import pandas as pd
from glob import glob
```

```python
# 파일 목록 한 번에 가져오기
glob("../data/지역_*.xls")
```

```python
# 파일명 저장
stations_files = glob("../data/지역_*.xls")
stations_files[:5]
```

```python
# 하나만 읽어보기
tmp = pd.read_excel(stations_files[0], header=2)
tmp.head(2)
```

```python
tmp_raw = []

for file_name in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
```

- 형식이 동일하고 연달아 붙이기만 하면 될 때는 concat

```python
station_raw = pd.concat(tmp_raw)
station_raw
```

```python
station_raw.info()
```

```python
station_raw.columns
```

```python
stations = pd.DataFrame({
    "상호" : station_raw["상호"],
    "주소" : station_raw["주소"],
    "가격" : station_raw["휘발유"],
    "셀프" : station_raw["셀프여부"],
    "상표" : station_raw["상표"]
})
stations.head()
```

```python
for eachAddress in stations["주소"]:
    print(eachAddress.split()[1])
```

```python
stations["구"] = [eachAddress.split()[1] for eachAddress in stations["주소"]]
stations.head()
```

```python
stations["구"].unique(), len(stations["구"].unique())
```

```python
# 가격 데이터형 변환 object => float

stations["가격"] = stations["가격"].astype("float")
```

```python
# 가격 정보 없는 주유소
stations[stations["가격"] == "-"]
```

```python
# 가격 정보 있는 주유소만 사용
stations[stations["가격"] != "-"]
stations.tail()
```

```python
stations["가격"] = stations["가격"].astype("float")
```

```python
stations.info()
```

```python
stations
```

```python
# 인덱스 재정렬

stations.reset_index(inplace=True)
stations.tail()
```

```python
del stations["index"]
stations
```

```python
import numpy as np

gu_data = pd.pivot_table(data=stations, index="구", values="가격", aggfunc=np.mean)
gu_data
```

```python
geo_path = "../data/02. skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding="utf-8"))

my_map = folium.Map(location=[37.5502, 126.982], zoom_start=10.5, tiles="OpenStreetMap")
folium.Choropleth(
    geo_data = geo_str,
    data = gu_data,
    columns = [gu_data.index, "가격"],
    key_on = "feature.id",
    fill_color="PuRd"
).add_to(my_map)
my_map
```
