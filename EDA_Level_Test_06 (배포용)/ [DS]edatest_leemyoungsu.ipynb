````python
from sklearn.linear_model import LinearRegression
# ê²°ê³¼ DataFrame ìƒì„±
result_data = []

for group in non_null_df['êµ¬ë¶„2'].unique():
    group_data = non_null_df[non_null_df['êµ¬ë¶„2'] == group]
    
    if len(group_data) > 1:
        # ê¸°ê°„ì— ë”°ë¥¸ íšŒê·€ê³„ìˆ˜ ê³„ì‚°
        X = group_data['ê¸°ê°„'].values.reshape(-1, 1)
        y = group_data['ì—¬ìë¹„ìœ¨'].values
        model = LinearRegression().fit(X, y)
        coef = round(model.coef_[0], 6)
        
        # MinMax ê³„ì‚°
        min_max = round(abs(np.max(y) - np.min(y)), 6)
        
        # í‰ê·  ê³„ì‚°
        mean = round(np.mean(y), 6)
        
        # í‘œì¤€í¸ì°¨ ê³„ì‚°
        std = round(np.std(y), 6)
        
        result_data.append([group, coef, min_max, mean, std])
result_data
````
